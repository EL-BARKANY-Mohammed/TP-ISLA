## ▶️ Exercise 3: Principal components analysis (5 points)

**(a)** Explain in a few words the main concepts and ideas underlying the principal component analysis (PCA). You should include both the geometric and statistical interpretations of PCA.

Principal Component Analysis (PCA) is a dimensionality reduction method. Its goal is to replace a large number of correlated variables by a smaller number of new variables, called principal components, while preserving as much information as possible.

Geometric interpretation:
Geometrically, PCA finds new orthogonal axes in the data space such that:
- the first principal component is the direction along which the projected data have the largest variance,
- the second principal component is the orthogonal direction with the next largest variance,
and so on.
- So PCA rotates the coordinate system to align it with the main directions of spread of the cloud of points.

Statistical interpretation:
Statistically, PCA builds new variables that are:
- linear combinations of the original variables,
- uncorrelated with each other,
- ordered by decreasing explained variance.

The first principal components summarize most of the variability of the data. Therefore, PCA is useful for reducing the dimension while keeping the essential structure of the dataset.

## ▶️ Exercise 5: PCR and cross-validation (6 points)

**(a)** Recall in a few words the principle of cross-validation. Explain why this procedure is useful when building a predictive model. Your answer should mention different strategies to handle datasets in which the samples are not IID.

Cross-validation is a method used to estimate the prediction performance of a model on unseen data. The dataset is split into several parts called folds. The model is trained on some folds and tested on the remaining fold, and this is repeated several times by changing the test fold. The final error is obtained by averaging the errors over all folds.

This is useful because it gives a more realistic estimate of generalization error than evaluating the model on the training data only.
When samples are not IID, the folds must be chosen carefully. For example:
- for time series, one should preserve temporal order,
- for grouped data, one should keep samples from the same group together,
- for correlated or repeated samples, one should avoid putting related observations in both train and test sets.

**(b)** Based on the structure of the dataset being used, such as the different countries of the individuals and the order in which the rows of the dataframe are provided, explain which choice of cross-validation iterator from [here](https://scikit-learn.org/stable/modules/cross_validation.html#cross-validation-iterators) seems the most adequate for our context.

In this dataset, individuals are grouped by population or tribe, and the rows are ordered by groups. Therefore, a standard random K-fold may create leakage because individuals from the same tribe could be in both train and test folds.

A more adequate choice is GroupKFold, using the tribe or population as the grouping variable. This keeps all individuals from the same tribe in the same fold and gives a more realistic estimate of prediction performance on unseen groups.


**(c)** We first assess the quality of the PCR fit for `n_components=4`. Note that you should be careful in avoiding [data leakage](https://scikit-learn.org/stable/common_pitfalls.html#data-leakage) problems when doing the PCA followed by a multiple linear regression. You should use the pipeline interface from scikit-learn with `sklearn.pipeline.make_pipeline` to facilitate your task. Be sure to evaluate the errors as done in **Exercise 4(b)**.

````python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import KFold

y= df.iloc[:, [6, 7]]
X = df.iloc[:, 8:]

pca=PCA(n_components=4)
lr=LinearRegression()

est=make_pipeline(pca,lr)
cv = KFold(n_splits=10, shuffle=True, random_state=0)
scores=[]
dists=[]

for train_idx, test_idx in cv.split(X, y):
    X_train, y_train=X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]

    est.fit(X_train, y_train)
    scores.append(est.score(X_test, y_test))
    y_pred=est.predict(X_test)
    y_true = y_test.to_numpy().reshape(-1, 2)
    y_pred = y_pred.reshape(-1, 2)
    
    # conversion degrés -> radians
    y_true_rad = np.radians(y_true)
    y_pred_rad = np.radians(y_pred)
    
    # distance 
    dist_matrix = haversine_distances(y_pred_rad, y_true_rad)
    error_km = 6371 * np.mean(np.diag(dist_matrix))
    dists.append(error_km)

    
print(np.mean(scores))
print(np.mean(dists))
````

Using a pipeline avoids leakage because PCA is recomputed separately inside each training fold. The test error is the relevant quantity for evaluating predictive performance.

**(d)** Repeat the analysis from item **(b)** but changing `n_components` between 2 and 440 in steps of 10. Plot the mean training and test errors versus the number of principal components. Attention, the errors should be given in kilometers.
